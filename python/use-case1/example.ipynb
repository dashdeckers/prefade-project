# Using prefade from Subdirectories

This notebook demonstrates how to use the `prefade` package from within the `python/notebooks/` subdirectory.

## Key Points:
- The `prefade` package is installed in editable mode in the virtual environment
- Any subdirectory within `python/` can import it normally
- No special path manipulation needed

```python
# Import the prefade module (just works!)
import prefade
from prefade import Email, NonEmptyStr, send_email


# Check what's available
print("Available from prefade:")
print(f"  - {prefade.Email}")
print(f"  - {prefade.NonEmptyStr}")
print(f"  - {prefade.send_email}")


def send_welcome(address: str) -> None:
    email = Email(address)          # validated in Rust
    subject = NonEmptyStr("Welcome!")
    body = NonEmptyStr("Thanks for joining.")
    send_email(email, subject, body)
```

```python
# Test the function
send_welcome("user@example.com")
```
